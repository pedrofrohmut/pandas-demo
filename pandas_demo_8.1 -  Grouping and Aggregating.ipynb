```python
# Str contains does not work directly on the SeriesGroupBy Object with have to use apply here
country_group["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
# 1. From country groups get the LanguageWorkedWith column values
# 2. Apply to each row return the lambda function that checks if it contains Python
# 3. Give the total number of rows where contains Python is True
country_group["LanguageWorkedWith"].apply(lambda x : x.str.contains("Python")).sum()
```

```python
# Gives the sum for each country (Group)
country_group["LanguageWorkedWith"].apply(lambda x : x.str.contains("Python").sum())
```

```python
country_group["LanguageWorkedWith"].apply(lambda x : len(x))
```

```python
# 1. Define the total number of people for a Group
# 2. Define the amount of people that knows Python in that Group
# 3. Devide Knows Python by Total to get the percentage
def percentage_that_knows_python(languages):
    num_people = len(languages)
    num_people_knows_python = languages.str.contains("Python").sum()
    return num_people_knows_python / num_people


country_group["LanguageWorkedWith"].apply(percentage_that_knows_python)
```

```python
# Lambda version of people that knows python percentage by group
country_group["LanguageWorkedWith"].apply(lambda x : x.str.contains("Python").sum() / len(x))
```

```python
python_df = pd.DataFrame()
```

```python
python_df["NumPeople"] = country_group["LanguageWorkedWith"].apply(lambda x : len(x))
```

```python
python_df["NumKnowPython"] = country_group[
    "LanguageWorkedWith"].apply(lambda x : x.str.contains("Python").sum())
```

```python
python_df["PerctKnowPython"] = country_group[
    "LanguageWorkedWith"].apply(lambda x : x.str.contains("Python").sum() / len(x))
```

```python
country_respondents = df["Country"].value_counts()
country_respondents
```

```python
country_uses_python = country_group[
    "LanguageWorkedWith"].apply(lambda x : x.str.contains("Python").sum())
country_uses_python
```

```python
py_df = pd.concat([country_uses_python, country_respondents], axis = "columns")
py_df.rename(
    columns = { 
        "LanguageWorkedWith": "KnowPython", 
        "Country": "Total"
    }, 
    inplace = True)
```

```python
py_df["PercentKnowPython"] = country_group[
    "LanguageWorkedWith"].apply(lambda x : ( x.str.contains("Python").sum() / len(x) ) * 100 )
```
